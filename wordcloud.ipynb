```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
from nltk.corpus import stopwords
import pandas as pd 
```

```python
df=pd.read_csv('labelled-reviews.csv')
df=df.dropna()
```

```python
df
```

```python
set=stopwords.words('english')
```

```python
def remove_stopwords(text):
    text=text.split()
    text=[word for word in text if not word in set]
    return ' '.join(text)
```

```python
remove_stopwords(df['reviews'][0])
```

```python
df['review_without_stopwords']=df['reviews'].apply(lambda x:remove_stopwords(x))
```

```python
wordcloud=WordCloud(width=800,height=800,stopwords='english').generate(' '.join(df['review_without_stopwords']))
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
```
